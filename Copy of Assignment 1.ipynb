## Question 1

- `Take two integer numbers and return their product. If the product is greater than 1000, then return their sum if product is less than 1000 return product value only.`



```python
# Please write The solution Here for above question 1:
def product(a,b):
  c=print(a*b)
  return c
  if c>1000:
    return a+b
  else:
    return a*b







product(45,50)



```

```python
def product_or_sum(a, b):
    if a * b <= 1000:
        return a * b
    else:
        return a + b

# Example usage:
result1 = product_or_sum(20, 30)  # Product is 600 (less than 1000)
result2 = product_or_sum(50, 10)  # Product is 500 (less than 1000)

print("Result 1:", result1)  # Output: Result 1: 600
print("Result 2:", result2)  # Output: Result 2: 500
```

## question 5

### Given a two Python list. Iterate both lists simultaneously such that list1 should display item in original order and list2 in reverse order

    list1 = [10, 20, 30, 40]
    list2 = [100, 200, 300, 400]


    Expected output:
    -----------------------------
    10 400
    20 300
    30 200
    40 100

```python
# Please write The solution Here for above question 5:
list1 = [10, 20, 30, 40]
list2 = [100, 200, 300, 400]

for x ,y in zip(list1,list2[::-1]):
  print(x,y)


```
