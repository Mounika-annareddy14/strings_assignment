### Question 5: Write a Python program to check if two strings are anagrams.

- `sample input : `
    - `var1 = 'listen'`
    - `var2 = 'silent'`
- `sample output :`
    - `result : both are anagram because characters in var2 is in var1`

```python
# Please write The solution Here for above question 5:anagram

var1='listen'
c=var1.split()
print(c)
var2='silent'
d=var2.split()
print(d)
#here for list we have a sort hence we can use sort function
e=c.sort()
f=d.sort()
if e==f:
  print('both are anagrams because characters in var2 is in var1')
else:
  print('not a anagram')



```

### Question 6: Write a Python program to remove all the vowels from a string.

- `sample input : `
    - `var = 'python'`
- `sample output :`
    - `result : pythn , because vowels are [AEIOUaeiou]`

```python
# Please write The solution Here for above question 6:

var='python'
c='aeiouAEIOU'
r=''
for i in var:
  if i not in c:
    r=r+i
print(r)




```
